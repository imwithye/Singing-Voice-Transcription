````python
from music21 import midi


def midi2stream(fp):
    mf = midi.MidiFile()
    mf.open(fp)
    mf.read()
    mf.close()
    return midi.translate.midiFileToStream(mf)
````

````python
import os
from utils import VALID_DATASET_DIR

song_idx = 401
````

````python
correct_midi = os.path.join(VALID_DATASET_DIR, str(song_idx), "Vocals.mid")
correct_s = midi2stream(correct_midi)
correct_s.show()
````

````python
predict_midi = os.path.join(VALID_DATASET_DIR, str(song_idx), "Vocals_predicteffnet_10.mid")
predict_s = midi2stream(correct_midi)
predict_s.show()
````
